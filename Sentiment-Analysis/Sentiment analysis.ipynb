4.1 Cleaning

   - Removed HTML tags, punctuation, numbers, and stopwords; converted all text to lowercase.

```python
#import re
#import nltk
#from nltk.corpus import stopwords

# Download stopwords if not already done
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize (split into words)
    words = text.split()
    # Remove stopwords
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# Apply preprocessing to the Reduced_df
Reduced_df['cleaned_review'] = Reduced_df['review'].apply(preprocess_text)
```

4.2 Sample of Original vs. Cleaned Text

   - Displayed original vs. cleaned reviews to verify the changes.

```python
# View the original and cleaned review side-by-side
Reduced_df[['review', 'cleaned_review']].sample(5, random_state=42)
```

4.3 Tokenization

  - Split cleaned reviews into individual words (tokens).

```python
# Tokenize the cleaned text into words
Reduced_df['tokens'] = Reduced_df['cleaned_review'].apply(lambda x: x.split())
```

4.4 Lemmatization

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')  # For better lemmatization accuracy
```

```python
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize each token
Reduced_df['lemmatized_tokens'] = Reduced_df['tokens'].apply(
    lambda tokens: [lemmatizer.lemmatize(word) for word in tokens]
)
```

```python
Reduced_df['final_cleaned_review'] = Reduced_df['lemmatized_tokens'].apply(lambda x: ' '.join(x))
```

## **Step 6  Sentiment Classification (Logistic Regression)**

6.1 Convert Cleaned Text into TF-IDF Features

```python
tfidf = TfidfVectorizer(max_features=5000)  # You can adjust max_features as needed
X = tfidf.fit_transform(Reduced_df['final_cleaned_review'])
```
